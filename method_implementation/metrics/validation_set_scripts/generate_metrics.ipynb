### Compute metrics to ground truth and save them in a CSV.


```python
from MEDIcaTe.calculate_dice_haus import dice_from_nifti, hausdorff_distance_from_nifti
from MEDIcaTe.file_folder_ops import *
import pandas as pd
from multiprocessing import Pool
```

```python
path_folder_ground_truth = '/homes/kovacs/project_data/hnc-auto-contouring/hnc_rigs_all/data_nifty/t_r196/labels'
path_folder_pred_nnunet = '/homes/kovacs/project_data/hnc-auto-contouring/hnc_rigs_all/data_nifty/t_r196/predictions/nnunet/labels_nn_unet_predict'
path_folder_pred_deep_medic = '/homes/kovacs/project_data/hnc-auto-contouring/hnc_rigs_all/data_nifty/t_r196/predictions/deep_medic/labels_deep_medic_predict'
path_folder_pred_deep_medic_norm = '/homes/kovacs/project_data/hnc-auto-contouring/hnc_rigs_all/data_nifty/t_r196/predictions/deep_medic_norm/labels_deep_medic_predict_norm'
path_folder_pred_tureckova = '/homes/kovacs/project_data/hnc-auto-contouring/hnc_rigs_all/data_nifty/t_r196/predictions/tureckova/tureckova_vnet1_lowres'
```

```python
# compute metrics for nnunet and write to folder
metric = []
value = []
method = []
case_list = []
result_dest = '/homes/kovacs/project_data/hnc-auto-contouring/hnc_rigs_all/data_final_results'
for m in [path_folder_pred_nnunet, path_folder_pred_deep_medic, path_folder_pred_deep_medic_norm, path_folder_pred_tureckova]:
    try:
        for i, case in enumerate(listdir(path_folder_ground_truth)):
            if i > 1000000:
                break

            print(case)
            fpath_gt = join(path_folder_ground_truth, case)
            fpath_method = join(m, case)
            d = dice_from_nifti(fpath_gt,fpath_method)
            h = hausdorff_distance_from_nifti(fpath_gt,fpath_method)
            value.append(d)
            value.append(h)

            metric.append('dice')
            metric.append('haus')

            method.append(basename(dirname(m)))
            method.append(basename(dirname(m)))

            case_list.append(case[:-7])
            case_list.append(case[:-7])
            
    except ValueError:
        value.append('na')
        method.append(basename(dirname(m)))
        case_list.append(case[:-7])

#print(haus)
d_metrics = pd.DataFrame({'case':case_list,'value':value,'metric':metric,'method':method})
d_metrics.to_csv(join(result_dest,'metrics.csv'))
```
